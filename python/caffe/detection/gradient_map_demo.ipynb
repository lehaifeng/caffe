```python
import caffe, caffe.imagenet as imagenet
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.append('lib/python/caffe/imagenet/')
import power_wrapper as pw

classify_net = caffe.CaffeNet(
     'caffe/examples/imagenet_classify.prototxt',
     'caffe/alexnet_train_iter_470000')
classify_net.set_phase_test()
localize_net = caffe.CaffeNet(
     'caffe/examples/imagenet_localize.prototxt',
     'caffe/alexnet_train_iter_470000')
localize_net.set_phase_train()

mean = imagenet.IMAGENET_MEAN[:227,:227]
```

```python
img_cow = pw.format_image(pw.load_image('cow.bmp'), cropped_size=True)
img_cow_disp = (img_cow.swapaxes(0,1).swapaxes(1,2) + pw.CROPPED_IMAGENET_MEAN)[:,:,::-1].astype(uint8)
img_cat = pw.format_image(pw.load_image('cat.jpg'), cropped_size=True)
img_cat_disp = (img_cat.swapaxes(0,1).swapaxes(1,2) + pw.CROPPED_IMAGENET_MEAN)[:,:,::-1].astype(uint8)
```

```python
def compute_feats_grads(img, classifier=None, localizer=None):       
    # single image batch, imagenet 1k classes output
    num_input = 1
    num_output = 1000
    im_arr = np.ascontiguousarray([img], dtype=np.float32)
    im_dims = (1,3,227,227)
    
    # classify by forward pass in classifier net
    input_blobs = [im_arr]
    output_blobs = [np.empty((num_input, num_output, 1, 1), dtype=np.float32)]
    classifier.Forward(input_blobs, output_blobs)
    feats = output_blobs[0].flatten()

    # find top class activation
    top_class = np.argmax(feats) # todo handle ties (pick most frequent class?)
    
    # localize by pixelwise gradients w.r.t. top activation forward-backward pass in localizer net
    input_blobs = [im_arr,
                   np.array([top_class], dtype=np.float32).reshape(1,1,1,1)]
    localizer.Forward(input_blobs, [])
    gradients = [np.empty(im_dims, dtype=np.float32),
                 np.empty((1,1,1,1), dtype=np.float32)]
    localizer.Backward([], gradients)
    gradients = gradients[0][0]

    return {'feat': feats, 'grad': gradients, 'predict': top_class}

def show_gradient(img_display, gradient):
    from scipy.ndimage import filters
    scale = (gradient ** 2).sum(0)
    scale /= scale.max()
    scale **= 0.25
    scale = filters.gaussian_filter(scale, sigma=5)
    plt.subplot(1,2,1)
    plt.imshow(img_display)
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(scale)
    plt.axis('off')
    plt.figure()
    plt.imshow(np.clip((img_display - gradient.swapaxes(0,1).swapaxes(1,2) * 1000), 0, 1)[:, :, ::-1])
```

```python
result_cow = compute_feats_grads(img_cow, classifier=classify_net, localizer=localize_net)
show_gradient(img_cow_disp, result_cow['grad'])
```

```python
result_cat = compute_feats_grads(img_cat, classifier=classify_net, localizer=localize_net)
show_gradient(img_cat_disp, result_cat['grad'])
```
